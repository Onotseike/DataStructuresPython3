```python
#### Singly Linked List

class SNode:
    def __init__(self, data):
        self.data = data
        self.next_pointer = None

class SinglyLinkedList:
    def __init__(self):
        self.head = None

    def print_list(self):
        temp_node = self.head
        while temp_node:
            print(temp_node.data, "=>", temp_node.next_pointer)
            temp_node = temp_node.next_pointer

    def prepend(self, new_node :SNode):
        temp = self.head
        new_node.next_pointer = temp
        self.head = new_node

    def append(self, new_node : SNode):
        temp = self.head
        if self.head is None:
            self.head = new_node
            return
        if self.head.next_pointer is None :
            self.head.next_pointer = new_node
            return
        while temp.next_pointer:
            temp = temp.next_pointer
        temp.next_pointer = new_node

    def is_node_in_list(self, node : SNode):
        current_node = self.head
        while current_node:
            if current_node.data == node.data and current_node.next_pointer == node.next_pointer:
                return True
            current_node = current_node.next_pointer
        return False

    def insert_post(self, post_node : SNode, new_node : SNode):
        if  post_node is None:
            print("Node to insert after can't be NONE")
            return
        if self.is_node_in_list(post_node):
            new_node.next_pointer = post_node.next_pointer
            post_node.next_pointer = new_node
            return
        print("Node corresponding to post_node not in the linked list")


    def delete_node(self, del_node : SNode):
        if  del_node is None:
            print("Node to delete can't be NONE")
            return
        current_node = self.head
        prev_node = None
        while current_node:
            if current_node.data == del_node.data and current_node.next_pointer == del_node.next_pointer:
                prev_node.next_pointer = current_node.next_pointer
                break
            else:
                prev_node = current_node
                current_node = current_node.next_pointer
```

```python
singly_linked_list = SinglyLinkedList()

singly_linked_list.append(SNode(1))
second_node = SNode(2)
third_node = SNode(3)
fourth_node = SNode(4)

singly_linked_list.append(second_node)
singly_linked_list.append(third_node)
singly_linked_list.append(fourth_node)

#Print list
singly_linked_list.print_list()

singly_linked_list.prepend(SNode(0))

print()
singly_linked_list.print_list()

singly_linked_list.append(SNode(5))
print()
singly_linked_list.print_list()

singly_linked_list.insert_post(third_node, SNode(3.50))
print()
singly_linked_list.print_list()

node = SNode(3.50)
node.next_pointer = fourth_node
singly_linked_list.delete_node(node)
print()

singly_linked_list.print_list()
```

```python
### Doubly Linked List

class DNode:
    def __init__(self, data):
        self.data = data
        self.prev_pointer = None
        self.next_pointer = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None

    def print_list_forward(self):
        temp = self.head
        while temp:
            print(temp.prev_pointer," <== ",temp.data," ==> ", temp.next_pointer)
            temp = temp.next_pointer

    def print_list_backward(self):
        temp = self.tail
        while temp:
            print(temp.next_pointer," ==> ",temp.data," <== ", temp.prev_pointer)
            temp = temp.prev_pointer

    def prepend(self, new_node : DNode):
        if  self.head is None:
            self.head = new_node
            return
        self.head.prev_pointer = new_node
        temp = self.head
        new_node.next_pointer = temp
        self.head = new_node

    def append(self, new_node : DNode):
        if self.tail is None:
            if self.head is None:
                self.head = new_node
                self.tail = new_node
                return
        self.tail.next_pointer = new_node
        temp = self.tail
        new_node.prev_pointer = temp
        self.tail = new_node

    def is_node_in_list(self, node : DNode):
        current_node = self.head
        if (node.data, node.prev_pointer, node.next_pointer) == (self.tail.data, self.tail.prev_pointer, self.tail.next_pointer):
            return True
        while current_node:
            if current_node.data == node.data and current_node.next_pointer == node.next_pointer and current_node.prev_pointer == node.prev_pointer:
                return True
            current_node = current_node.next_pointer
        return False

    def insert_post_node(self, post_node : DNode, new_node : DNode):
        if post_node is None:
            print("Node cannot be None")
            return
        if(post_node.data, post_node.prev_pointer, post_node.next_pointer) == (self.tail.data, self.tail.prev_pointer, self.tail.next_pointer):
            self.append(new_node)
            return
        if self.is_node_in_list(post_node):
            new_node.prev_pointer = post_node
            new_node.next_pointer = post_node.next_pointer
            post_node.next_pointer = new_node
            new_node.next_pointer.prev_pointer = new_node
            return
        print("Node corresponding to the post_node is not in the list")

    def insert_pre_node(self, pre_node : DNode, new_node : DNode):
        if pre_node is None:
            print("Node cannot be None")
            return
        if(pre_node.data, pre_node.prev_pointer, pre_node.next_pointer) == (self.head.data, self.head.prev_pointer, self.head.next_pointer):
            self.prepend(new_node)
            return
        if self.is_node_in_list(pre_node):
            new_node.next_pointer = pre_node
            new_node.prev_pointer = pre_node.prev_pointer
            pre_node.prev_pointer = new_node
            return
        print("Node corresponding to the pre_node is not in the list")

    def delete_node(self, node : DNode):
        if node is None:
            print("Node cannot be null")
            return
        if self.is_node_in_list(node):
            temp = node
            node.prev_pointer.next_pointer = temp.next_pointer
            node.next_pointer.prev_pointer = temp.prev_pointer
            return
        print("Node is not in the list")
```

```python
doubly_linked_list = DoublyLinkedList()
doubly_linked_list.append(DNode(1))
doubly_linked_list.append(DNode(2))
doubly_linked_list.append(DNode(3))
doubly_linked_list.append(DNode(4))
doubly_linked_list.append(DNode(5))

print("Forward Print")
doubly_linked_list.print_list_forward()
print("Backward Print")
doubly_linked_list.print_list_backward()
print()

print("Insert after node")
doubly_linked_list.insert_post_node(doubly_linked_list.head, DNode(1.5))
print("Forward Print")
doubly_linked_list.print_list_forward()
print("Backward Print")
doubly_linked_list.print_list_backward()
print()

print("Insert before node")
doubly_linked_list.insert_pre_node(doubly_linked_list.head, DNode(0.5))
print("Forward Print")
doubly_linked_list.print_list_forward()
print("Backward Print")
doubly_linked_list.print_list_backward()
print()

print("Insert at the beginning of list")
doubly_linked_list.prepend(DNode(0))
print("Forward Print")
doubly_linked_list.print_list_forward()
print("Backward Print")
doubly_linked_list.print_list_backward()
print()

print("Insert at the end of list")
doubly_linked_list.append(DNode(6))
print("Forward Print")
doubly_linked_list.print_list_forward()
print("Backward Print")
doubly_linked_list.print_list_backward()
print()

print("Delete node")
point_five_node = DNode(0.5)
point_five_node.prev_pointer = doubly_linked_list.head
point_five_node.next_pointer = doubly_linked_list.head.next_pointer.next_pointer
doubly_linked_list.delete_node(point_five_node)
print("Forward Print")
doubly_linked_list.print_list_forward()
print("Backward Print")
doubly_linked_list.print_list_backward()
print()


print("Delete node")
one_point_five_node = DNode(1.5)
one_point_five_node.prev_pointer = doubly_linked_list.head.next_pointer
one_point_five_node.next_pointer = doubly_linked_list.head.next_pointer.next_pointer.next_pointer
doubly_linked_list.delete_node(one_point_five_node)
print("Forward Print")
doubly_linked_list.print_list_forward()
print("Backward Print")
doubly_linked_list.print_list_backward()
print()
```

```python
#### Circular Linked List

class CNode:
    def __init__(self, data):
        self.data = data
        self.next_pointer = None

class CircularLinkedList:
    def __init__(self):
        self.head = None

    def print_list(self):
        temp_node = self.head
        while temp_node:
            print(temp_node.data, "=>", temp_node.next_pointer)
            temp_node = temp_node.next_pointer

    def prepend(self, new_node :CNode):
        temp = self.head
        new_node.next_pointer = temp
        self.head = new_node

    def append(self, new_node : CNode):
        temp = self.head
        if self.head is None:
            self.head = new_node
            self.head.next_pointer = new_node
            return
        if self.head.next_pointer is None :
            self.head.next_pointer = new_node
            return
        while temp.next_pointer:
            temp = temp.next_pointer
        temp.next_pointer = new_node
        new_node.next_pointer = self.head

    def is_node_in_list(self, node : CNode):
        current_node = self.head
        while current_node:
            if current_node.data == node.data and current_node.next_pointer == node.next_pointer:
                return True
            current_node = current_node.next_pointer
        return False

    def insert_post(self, post_node : CNode, new_node : CNode):
        if  post_node is None:
            print("Node to insert after can't be NONE")
            return
        if self.is_node_in_list(post_node):
            new_node.next_pointer = post_node.next_pointer
            post_node.next_pointer = new_node
            return
        print("Node corresponding to post_node not in the linked list")


    def delete_node(self, del_node : CNode):
        if  del_node is None:
            print("Node to delete can't be NONE")
            return
        current_node = self.head
        prev_node = None
        while current_node:
            if current_node.data == del_node.data and current_node.next_pointer == del_node.next_pointer:
                prev_node.next_pointer = current_node.next_pointer
                break
            else:
                prev_node = current_node
                current_node = current_node.next_pointer
```
