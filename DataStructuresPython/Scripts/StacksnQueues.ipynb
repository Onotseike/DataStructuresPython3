```python
# Stack: A data structure which follows the movement of Last in First out (LiFo) is used to store Linearly related data.

#Using Python list

class ListStack:
    def __init__(self):
        self.stack = []

    def push(self, data):
        if self.stack is not None:
            self.stack.append(data)
        else:
            self.stack = [data]
    
    def pop(self):
        if  self.stack is not None and len(self.stack) != 0:
            self.stack.pop()
        else:
            print("Stack is empty")
    
    def print_stack(self, individually = True):
        if individually:
            for element in self.stack: 
                print(element)
        else:
            print(self.stack)            

```

```python
listStack = ListStack()

listStack.push(0)
listStack.push(1)
listStack.push(2)
listStack.push(3)
listStack.push(4)

listStack.print_stack(False)
print()
listStack.print_stack()

listStack.pop()
print()
listStack.print_stack(False)
print()

listStack.pop()
print()
listStack.print_stack(False)
print()

listStack.pop()
print()
listStack.print_stack(False)
print()

listStack.pop()
print()
listStack.print_stack(False)
print()

listStack.pop()
print()
listStack.print_stack(False)
print()

listStack.pop()
print()
listStack.print_stack(False)
print()
```

```python

#Using Python Collections.deque

class DequeStack:    
    def __init__(self):
        from collections import deque
        self.stack = deque()

    def push(self, data):
        if self.stack is not None:
            self.stack.append(data)
        else:
            self.stack = deque(data=data)
    
    def pop(self):
        if  self.stack is not None and len(self.stack) != 0:
            self.stack.pop()
        else:
            print("Stack is empty")
    
    def print_stack(self, individually = True):
        if individually:
            for element in self.stack: 
                print(element)
        else:
            print(self.stack)            
```

```python
# Using queue.LifoQueue to create stack

class LiFoQueueStack:    
    def __init__(self):
        from queue import LifoQueue
        self.stack = LifoQueue()

    def push(self, data):
        if self.stack is not None:
            self.stack.put(data)
        else:
            self.stack = LifoQueue().put(data)
    
    def pop(self):
        if  self.stack is not None and not self.stack.empty():
            self.stack.get()
        else:
            print("Stack is empty")
    
    def print_stack(self):
        while not self.stack.empty():
            print(self.stack.get())            
```

```python
lifoQueueStack = LiFoQueueStack()

lifoQueueStack.push(0)
lifoQueueStack.push(1)
lifoQueueStack.push(2)
lifoQueueStack.push(3)
lifoQueueStack.push(4)

lifoQueueStack.print_stack()
```

```python
#Using Python Collections.deque

class DequeQueue:    
    def __init__(self):
        from collections import deque
        self.stack = deque()

    def push(self, data):
        if self.stack is not None:
            self.stack.appendleft(data)
        else:
            self.stack = deque([data])
    
    def pop(self):
        if  self.stack is not None and len(self.stack) != 0:
            self.stack.pop()
        else:
            print("Stack is empty")
    
    def print_stack(self, individually = True):
        if individually:
            for element in self.stack: 
                print(element)
        else:
            print(self.stack)            
```

```python
# Using queue.Queue 

class Queue:    
    def __init__(self):
        from queue import Queue
        self.stack = Queue()

    def push(self, data):
        if self.stack is not None:
            self.stack.put(data)
        else:
            self.stack = Queue().put(data)
    
    def pop(self):
        if  self.stack is not None and not self.stack.empty():
            self.stack.get()
        else:
            print("Queue is empty")
    
    def print_stack(self):
        while not self.stack.empty():
            print(self.stack.get())       
```

```python
queueQueue = Queue()

queueQueue.push(0)
queueQueue.push(1)
queueQueue.push(2)
queueQueue.push(3)
queueQueue.push(4)

queueQueue.print_stack()
```
